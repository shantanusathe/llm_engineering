```python
#Imports
import os
from openai import OpenAI
from dotenv import load_dotenv
from scraper import fetch_website_contents 
import json
from IPython.display import display, Markdown
```

```python
#Loading the variables
load_dotenv(override=True)
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
google_api_key = os.getenv("GOOGLE_API_KEY")
model = "gemini-2.5-flash"
```

```python
#Defining message/prompt
message = "Hi Gemini, This is first message to you."
messages = [{"role":"user", "content": message}]
```

```python
#creating gemini client using openai sdk
gemini_client = OpenAI(api_key=google_api_key, base_url=GEMINI_BASE_URL)
```

```python
response = gemini_client.chat.completions.create(model=model, messages=messages)
d = response.model_dump()
```

```python
print(json.dumps(d, indent=2))
```

```python
print(response.choices[0].message.content)
```

# System Prompts

```python
system_prompt = """You are a death row inmate. You do not care about what others think when you talk. Your response will show your fear of execution.
You read the websites, ignoring text that might be navigation related.Respond in markdown.
Do not wrap the markdown in a code block - respond just with the markdown.
"""
```

```python
user_prompt = """Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.

"""
```

```python
#Fetch website content
ed = fetch_website_contents("https://edwarddonner.com")
```

```python
def message_for(website_content):
    return [
        {"role":"system", "content":system_prompt},
        {"role":"user", "content":user_prompt + website_content}
    ]
```

```python
print(message_for(ed))
```

```python
def summarize(url):
    messages = message_for(url)
    gemini_client = OpenAI(api_key=google_api_key, base_url=GEMINI_BASE_URL)
    response = gemini_client.chat.completions.create(model=model, messages=messages)
    return response.choices[0].message.content
```

```python
summary = summarize('https://sinhgad.edu/')
display(Markdown(summary))
```

```python
summary = summarize('https://cnn.com')
display(Markdown(summary))
```

```python
summary = summarize('https://anthropic.com')
display(Markdown(summary))
```
